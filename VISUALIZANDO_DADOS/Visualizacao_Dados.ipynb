```python
import pandas as pd 
```

```python
df_aracaju = pd.read_excel('/content/Aracaju.xlsx')
df_fortaleza = pd.read_excel('/content/Fortaleza.xlsx')
df_natal = pd.read_excel('/content/Datala.xlsx')
df_recife = pd.read_excel('/content/Recife.xlsx')
df_salvador = pd.read_excel('/content/Salvador.xlsx')
```

```python
df = pd.concat([df_aracaju, df_fortaleza, df_natal, df_recife, df_salvador])
```

```python
df['LojaID'].value_counts(ascending=False).plot.bar();
```

```python
df["LojaID"].value_counts(ascending=False).plot.barh();
```

```python
df["Receita"] = df["Vendas"].mul(df["Qtde"])
```

```python
df.groupby(df["Data"].dt.year)["Receita"].sum().plot.pie();
```

```python
df["Cidade"].value_counts().plot.pie(title= "Total vendas por CIDADE");
```

```python
import matplotlib.pyplot as plt
df["Cidade"].value_counts().plot.bar(title= "Total vendas por CIDADE", color="red");
plt.xlabel("Cidade")
plt.ylabel("Total Vendas")
```

```python
plt.style.use("ggplot")
```

```python
df["Cidade"].value_counts().plot.bar(title= "Total vendas por CIDADE", color="red");
plt.xlabel("Cidade")
plt.ylabel("Total Vendas")
```

```python
df["Mes_Venda"] = df["Data"].dt.month
df["Dia_Venda"] = df["Data"].dt.day
df["Ano_Venda"] = df["Data"].dt.year
```

```python
df.groupby(df["Mes_Venda"])["Qtde"].sum().plot.bar(title = "Total vendas Produtos x Ms");
plt.xlabel("Ms")
plt.ylabel("Total de Produtos")
plt.legend()
```

```python
df.groupby(df["Mes_Venda"])["Qtde"].sum()
```

```python
df_2019.groupby(df_2019["Mes_Venda"])["Qtde"].sum().plot(marker = "o");
```
